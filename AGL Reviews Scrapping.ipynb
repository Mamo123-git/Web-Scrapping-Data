```python
# get html
import urllib
import urllib.request

from bs4 import BeautifulSoup
```

```python
page_list = []
def pagination():
    #for page_link in soup.find_all('a', {'class': 'pagination-page'}):
    for i in range(27,33):
        domain = "https://www.trustpilot.com/review/agl.com.au?page="+str(i)
        page_list.append(domain)

    return page_list
pages = pagination()

print(pages)
```

```python
url_list = []
def get_urls(soup):
    for link in soup.find_all('a', {'class': 'link link--large link--dark'}):
        domain = "https://www.trustpilot.com/"
        url = link.get('href')
        url_list.append(domain+url)
    return list(set(url_list))
```

```python
def get_page(url):
    fp = urllib.request.urlopen(url)
    mybytes = fp.read()
    mystr = mybytes.decode("utf8")
    fp.close()
    
    html_doc = mystr
    soup = BeautifulSoup(html_doc, 'html.parser')
    return soup
```

```python
urls = [get_urls(get_page(page)) for page in pages]
urls
```

```python
from pandas.core.common import flatten
flat_urls = list(flatten(urls))
```

```python
len(flat_urls)
```

```python
import time
comment_list = []
def get_comments(urls):
    for url in urls:
        try:
            print(url)
            soup = get_page(url)
            comment = soup.find('p', {'class':'review-content__text'}).get_text(strip=True)
            comment_list.append(comment)
        except:
            continue
        #time.sleep(30)
    return comment_list
comments = get_comments(flat_urls)
```

```python
df = pd.DataFrame(comment_list, columns=['comments'])
df.shape
```
